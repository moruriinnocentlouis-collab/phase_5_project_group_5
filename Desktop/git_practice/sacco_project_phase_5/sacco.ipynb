## Step 9: Cleaning the Date Column

The Date column contains inconsistent formats, including:
- Year-based values (e.g. Aud-2019, audited 2022)
- Month names (e.g. Jan, Feb)
- Missing values

To make the dataset usable, we standardize the Date column into a consistent format.

This step is essential for time-based analysis and modeling.

````python
import numpy as np

def clean_date(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip().lower()
    
    if "2019" in x:
        return "2019"
    elif "2020" in x:
        return "2020"
    elif "21" in x:
        return "2021"
    elif "2022" in x:
        return "2022"
    else:
        return x.capitalize()  # Jan, Feb, etc.

df_long['Clean_Date'] = df_long['Date'].apply(clean_date)

df_long[['Date', 'Clean_Date']].drop_duplicates()
````
