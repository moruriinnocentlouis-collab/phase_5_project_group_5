## Step 4: Understanding Data Structure

After cleaning the header rows, the dataset now shows financial line items across multiple time periods.

- Columns represent different time periods (years and months)
- The dataset is in a wide format

- Rename key columns
- Reshape the dataset into a structured format suitable for analysis

```python
df = df.rename(columns={
    'Unnamed: 0': 'Account_ID',
    'FINANCIAL YEAR 2021': 'Account_Name'
})

df.head()
```

```python
## Step 5: Cleaning Column Names

Some column names contained hidden spaces, which prevented proper renaming.

We cleaned the column names by removing extra spaces and then renamed key columns:
- Account_ID
- Account_Name

This ensures consistency and avoids errors in later steps.
```

```python
#renaming again
df = df.rename(columns={
    'Unnamed: 0': 'Account_ID',
    'FINANCIAL YEAR  2021': 'Account_Name'
})
```

```python
#checking again
df.head()
```

## Step 6: Removing Irrelevant Columns

The dataset contains several empty or irrelevant columns labeled as "Unnamed".

These columns do not provide useful information for analysis and may introduce errors.

We remove these columns to simplify the dataset and focus only on meaningful data.

```python
df.columns.str.contains('Unnamed')
```

```python
df = df.loc[:, ~df.columns.astype(str).str.contains('Unnamed')]

df.head()
```

## Step 7: Reshaping the Dataset (Wide to Long Format)

The dataset is currently in a wide format where each time period (months and years) is represented as a separate column.

This structure is not suitable for analysis or modeling.

To make the data usable, we convert it into a long format where:
- Each row represents a single observation
- Time periods are stored in one column
- Values are stored in another column

This transformation is essential for time series analysis and predictive modeling.

```python
df_long = df.melt(
    id_vars=['Account_ID', 'Account_Name'],
    var_name='Date',
    value_name='Value'
)

df_long.head()
```

## Step 8: Structured Dataset Achieved

The dataset has now been transformed into a long format.

Each row represents:
- A financial category (Account_Name)
- A specific time period (Date)
- A corresponding value (Value)

This structure allows us to:
- Analyze trends over time
- Compare financial categories
- Prepare the data for feature selection and modeling

```python
df_long['Date'].unique()
```

## Step 9: Cleaning the Date Column

The Date column contains inconsistent formats, including:
- Year-based values (e.g. Aud-2019, audited 2022)
- Month names (e.g. Jan, Feb)
- Missing values

To make the dataset usable, we standardize the Date column into a consistent format.

This step is essential for time-based analysis and modeling.

```python
import numpy as np

def clean_date(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip().lower()
    
    if "2019" in x:
        return "2019"
    elif "2020" in x:
        return "2020"
    elif "21" in x:
        return "2021"
    elif "2022" in x:
        return "2022"
    else:
        return x.capitalize()  # Jan, Feb, etc.

df_long['Clean_Date'] = df_long['Date'].apply(clean_date)

df_long[['Date', 'Clean_Date']].drop_duplicates()
```

## Step 10: Removing Invalid Dates

Some rows contain missing date values.

These rows do not contribute to time-based analysis and may introduce errors.

We remove them to ensure the dataset is clean and consistent.

```python
df_long = df_long.dropna(subset=['Clean_Date'])

df_long['Clean_Date'].unique()
```

## Step 11: Creating Features from Financial Categories

Currently, financial categories (Account_Name) exist as rows.

To prepare for analysis and modeling, we convert these into columns.

This transformation creates a structured dataset where:
- Each row represents a time period
- Each column represents a financial feature

```python
df_pivot = df_long.pivot_table(
    index='Clean_Date',
    columns='Account_Name',
    values='Value',
    aggfunc='sum'
)

df_pivot.head()
```

## Step 12: Final Structured Dataset

The dataset has been successfully transformed into a structured format suitable for analysis.

Key characteristics:
- Each row represents a time period
- Each column represents a financial variable (feature)
- Values represent financial amounts

This structure allows for:
- Time series analysis
- Feature selection
- Predictive modeling

This marks the transition from data preparation to analytical modeling.

## Step 13: Filtering Annual Data

The dataset contains both monthly and yearly data.

To simplify analysis and ensure consistency, we focus on annual data only.

This reduces noise and allows for clearer trend analysis and modeling.

```python
df_pivot = df_pivot.loc[['2019', '2020', '2021', '2022']]

df_pivot
```

## Step 14: Feature Selection 

Feature selection is based on actual available financial variables in the dataset.

We select key variables that represent:
- Financial strength
- Lending activity
- Liquidity
- Capital structure

These features are important for predicting SACCO performance.

```python
#Step 15-Creating model dataset
features = [
    'Total Assets',
    'Total Deposit liabilities',
    'Total Equity',
    'Total Liabilities',
    'Cash & Cash Equivalent',
    'Financial Investments',
    'Gross Loan Portfolio',
    'Net Loan Portfolio',
    'Non withdrawable deposits (BOSA member deposits)',
    'Share Capital (Members share capital per Bylaws)',
    'Statutory Reserve'
]

target = "Current Year's Surplus"
```

```python
list(df_pivot.columns)
```

```python
#Build Model Dataset.
df_model = df_pivot[features + [target]]

df_model
```
